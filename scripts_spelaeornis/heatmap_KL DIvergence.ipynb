```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.colors as colors
```

```python
df=pd.read_csv("Kl divergence.csv", header=0, index_col=0)
df=pd.DataFrame(df)
vmin=np.nanmin(df)
vmax=np.nanmax(df)
mask=df.isnull()
plt.figure(figsize=(15,10))

heatmap=sns.heatmap(data=df, vmin=vmin-100, vmax=vmax+100, annot=True, annot_kws={"fontsize":18},fmt='.5g')
plt.yticks(rotation=0)
for label in heatmap.get_yticklabels():
        # Here it sets all italic. 
        label.set_style("italic")
for label in heatmap.get_xticklabels():
        # Here it sets all italic. 
        label.set_style("italic")
sns.set(font_scale=2)
plt.title('Pair-wise statistic of Kullback-Leibler divergence ')
plt.savefig('KL divergence heatmap.svg')

```
